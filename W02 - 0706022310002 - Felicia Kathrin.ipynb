```python
import pandas as pd

# Download the CSV file from the URL
url = "https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv"
df = pd.read_csv(url, sep=',')

df.head()

# Melakukan load dataset using pandas dataframe dan menampilkan 5 rows pertama data
```

```python
df = df.drop(['url', 'ads_id'], axis=1)
df.head()

# Selajutnya melakukan drop data, menurut saya kolom url dan ads_id tidak bisa digunakan untuk analisis,
# sehingga saya memilih untuk membuang semua data di kolom tersebut.
```

```python
# Calculate null values and percentage
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]


# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)


null_df

# Lalu saya ingin mengetahui jumlah data yang null di setiap kolom beserta persentase nya
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the horizontal bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.barplot(x='Null Count', y=null_df.index, data=null_df, orient='h', palette='rainbow')
plt.xlabel('Number of Null Values', fontsize=12)
plt.ylabel('Columns', fontsize=12)
plt.title('Null Values per Column', fontsize=14)

# Add labels to the bars
for i, v in enumerate(null_df['Null Count']):
    plt.text(v + 1, i, str(v), color='black', va='center', fontsize=10)

plt.tight_layout()
plt.show()

# ini visualisasi null value setiap kolom
```

```python
# Dari 2 bar plot diatas, dapat diketahui bahwa ada data yang tertukar antara baru di furnishing, lalu unfurnished semi furnished di property condition.
# Seharusnya data baru ada di kolom property condition dan furnished semi furnished di kolom furnishing
# Sehingga dibawah ini akan dibenarkan

# Jika property_condition salah (berisi kata furnishing)
mask1 = df['property_condition'].isin(['unfurnished', 'semi furnished', 'furnished'])
df.loc[mask1, ['property_condition', 'furnishing']] = df.loc[mask1, ['furnishing', 'property_condition']].values

# Jika furnishing salah (berisi kata condition)
mask2 = df['furnishing'].isin(['bagus', 'baru', 'bagus sekali', 'sudah renovasi', 'butuh renovasi'])  # tambahkan kategori condition lain kalau ada
df.loc[mask2, ['property_condition', 'furnishing']] = df.loc[mask2, ['furnishing', 'property_condition']].values
```

```python
df['building_orientation'].fillna('Unknown', inplace=True)
df['certificate'].fillna('Unknown', inplace=True)
df['furnishing'].fillna('Unknown', inplace=True)
df['property_condition'].fillna('Unknown', inplace=True)

# Untuk keempat kolom diatas, diputuskan untuk mengisi null value dengan unknown.
# Dikarenakan termasuk kategorikal data serta menjadi cara yang aman untuk tidak
# mengubah data dengan input yang berbeda
```

```python
df['building_age'].fillna(df['building_age'].median(), inplace=True)
df['year_built'].fillna(df['year_built'].median(), inplace=True)

# building_age dan year_built juga memiliki null value yang cukup banyak
# Dan kedua kolom tersebut null pada rows data yang sama, sehingga tidak dapat
# memakai salah satu kolom saja untuk mengetahui sudah berapa lama bangunan berdiri.
# Sehingga lebih baik diisi dengan median
```

```python
# Selajutnya melakukan binning pada building_age. Agar lebih mudah untuk
# mengenali apakah bangunan ini termasuk baru ataupun tua
bins = [0, 10, 20, 100]  # batas umur bangunan
labels = ['New', 'Moderate', 'Old']  # label

df['building_age_bin'] = pd.cut(df['building_age'], bins=bins, labels=labels, right=True, include_lowest=True)

# cek hasil
print(df[['building_age', 'building_age_bin']].head(10))
```

```python
# Melakukan binning pada data price untuk mengelompokkan rumah yang mahal, medium, dan murah.
# Serta dapat mempermudah analisis terakhir

bins = [0, 2000000000, 3000000000, 4000000000, df['price_in_rp'].max()]
labels = ['Low', 'Medium', 'High', 'Luxury']

df['price_bin'] = pd.cut(df['price_in_rp'], bins=bins, labels=labels, include_lowest=True)

print(df[['price_in_rp', 'price_bin']].head(10))
```
